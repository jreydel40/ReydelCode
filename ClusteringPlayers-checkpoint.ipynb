```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Filter full_data to include only second basemen (2B)


# Columns of interest
include_columns_start = [
    'Name', 'PlayerID', 'Season', 'Age', 'G', 'PA', 'H',
    'B1', 'B2', 'B3', 'HR', 'BB', 'SO', 'AVG', 'OBP', 'SLG', 'TB', 'HBP',
    'wRCPlus', 'Bats', 'Throws', 'Height', 'Weight', 'Position'
]

# Select only the specified columns
full_start = full_data.filter(include_columns_start)

# Label encoding for categorical variables
label_encoders = {}
for column in ['Bats', 'Throws', 'Position']:
    le = LabelEncoder()
    full_start[column] = le.fit_transform(full_start[column])
    label_encoders[column] = le

# Select only numerical columns for clustering
numerical_columns = full_start.drop(['Name', 'PlayerID', 'Season', 'Position', 'Height'], axis=1)

# Check for missing values and handle them if any
numerical_columns = numerical_columns.dropna()

# Standardize the data
scaler = StandardScaler()
full_scaled = scaler.fit_transform(numerical_columns)

# Initialize an empty list to store within-cluster sum of squares (WCSS)
wcss = []

# Try KMeans for different values of k from 1 to 10
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(full_scaled)
    # Append the WCSS for the current k to the list
    wcss.append(kmeans.inertia_)

# Plot the elbow curve
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.title('Elbow Method for Optimal k')
plt.xticks(range(1, 11))
plt.grid(True)
plt.show()

# Assuming the optimal k is chosen from the elbow method (let's say k=4 for this example)
optimal_k = 4  # Replace with the actual number determined from the elbow plot
full_kmeans = KMeans(n_clusters=optimal_k, random_state=42)
full_start['Cluster'] = full_kmeans.fit_predict(full_scaled)

# Plotting the clustering results
plt.figure(figsize=(12, 8))

# Create a scatter plot of PA vs. wRCPlus with cluster colors
for cluster in full_start['Cluster'].unique():
    clustered_data = full_start[full_start['Cluster'] == cluster]
    plt.scatter(clustered_data['PA'], clustered_data['wRCPlus'], label=f'Cluster {cluster}', alpha=0.6)

# Adding titles and labels
plt.title('Clustering of All based on PA and wRCPlus')
plt.xlabel('Plate Appearances (PA)')
plt.ylabel('Weighted Runs Created Plus (wRCPlus)')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()

# Display the cluster assignments with name, season, and stats
columns_of_interest = [
    'Name', 'PlayerID', 'Season', 'Age', 'G', 'PA', 'H',
    'B1', 'B2', 'B3', 'HR', 'BB', 'SO', 'AVG', 'OBP', 'SLG', 'TB', 'HBP',
    'wRCPlus', 'Bats', 'Throws', 'Height', 'Weight', 'Position', 'Cluster'
]
# Filter the B2_start to include only these columns and sort by cluster in descending order
all_clusters = full_start[columns_of_interest].sort_values(by='Cluster', ascending=False)

# Display the resulting DataFrame
print(all_clusters)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
from scipy.spatial.distance import cdist

# Function to preprocess data
def preprocess_data(data, columns_of_interest, label_encoders=None):
    data_filtered = data[columns_of_interest].copy()
    
    # Label encoding for categorical variables
    if label_encoders is None:
        label_encoders = {}
        for column in ['Bats', 'Throws', 'Position']:
            le = LabelEncoder()
            data_filtered[column] = le.fit_transform(data_filtered[column])
            label_encoders[column] = le
    else:
        for column in ['Bats', 'Throws', 'Position']:
            le = label_encoders[column]
            data_filtered[column] = le.transform(data_filtered[column])
    
    # Select only numerical columns for clustering
    numerical_columns = data_filtered.drop(['Name', 'PlayerID', 'Season', 'Position', 'Height'], axis=1)
    
    # Check for missing values and handle them if any
    numerical_columns = numerical_columns.dropna()
    
    # Standardize the data
    scaler = StandardScaler()
    data_scaled = scaler.fit_transform(numerical_columns)
    
    return data_filtered, data_scaled, label_encoders

# Columns of interest
include_columns_start = [
    'Name', 'PlayerID', 'Season', 'Age', 'G', 'PA', 'H',
    'B1', 'B2', 'B3', 'HR', 'BB', 'SO', 'AVG', 'OBP', 'SLG', 'TB', 'HBP',
    'wRCPlus', 'Bats', 'Throws', 'Height', 'Weight', 'Position'
]

# Filter full_data to include only shortstops (SS) and second basemen (2B)
ss_data = full_data[full_data['Position'] == 'SS']
b2_data = full_data[full_data['Position'] == '2B']

# Combine the data to fit the label encoders on all possible values
combined_data = pd.concat([ss_data, b2_data])

# Preprocess the combined data to fit the label encoders
_, _, label_encoders = preprocess_data(combined_data, include_columns_start)

# Preprocess the data for SS and 2B separately using the fitted label encoders
ss_filtered, ss_scaled, _ = preprocess_data(ss_data, include_columns_start, label_encoders)
b2_filtered, b2_scaled, _ = preprocess_data(b2_data, include_columns_start, label_encoders)

# Perform clustering on SS data
optimal_k_ss = 10  # Replace with the actual number determined from the elbow plot
kmeans_ss = KMeans(n_clusters=optimal_k_ss, random_state=42)
ss_filtered['Cluster'] = kmeans_ss.fit_predict(ss_scaled)

# Perform clustering on B2 data
optimal_k_b2 = 10  # Replace with the actual number determined from the elbow plot
kmeans_b2 = KMeans(n_clusters=optimal_k_b2, random_state=42)
b2_filtered['Cluster'] = kmeans_b2.fit_predict(b2_scaled)

# Align clusters between SS and B2
ss_cluster_centers = kmeans_ss.cluster_centers_
b2_cluster_centers = kmeans_b2.cluster_centers_

# Compute the distance between each SS cluster center and each B2 cluster center
distances = cdist(ss_cluster_centers, b2_cluster_centers)

# Find the closest B2 cluster center for each SS cluster center
closest_clusters = np.argmin(distances, axis=1)

# Create a mapping from old B2 cluster labels to new SS-aligned cluster labels
cluster_mapping = {old: new for new, old in enumerate(closest_clusters)}

# Apply the mapping to the B2 clusters
b2_filtered['Aligned_Cluster'] = b2_filtered['Cluster'].map(cluster_mapping)
b2_filtered['Aligned_Cluster'].fillna(0, inplace=True)
b2_filtered['Aligned_Cluster'] = b2_filtered['Aligned_Cluster'].astype(int)  # Convert to integers



# Define custom colors for clusters
cluster_colors = {
    0: 'red', 
    1: 'green', 
    2: 'yellow', 
    3: 'blue', 
    4: 'pink', 
    5: 'orange', 
    6: 'black', 
    7: 'grey', 
    8: 'brown', 
    9: 'purple'
}

# Plotting the clustering results for SS
plt.figure(figsize=(12, 8))
for cluster in sorted(ss_filtered['Cluster'].unique()):
    clustered_data = ss_filtered[ss_filtered['Cluster'] == cluster]
    plt.scatter(clustered_data['PA'], clustered_data['wRCPlus'], label=f'Cluster {cluster}', c=cluster_colors[cluster], alpha=0.6)
plt.title('Clustering of SS based on PA and wRCPlus')
plt.xlabel('Plate Appearances (PA)')
plt.ylabel('Weighted Runs Created Plus (wRCPlus)')
plt.legend()
plt.grid(True)
plt.show()

# Plotting the clustering results for 2B with aligned cluster labels
plt.figure(figsize=(12, 8))
for cluster in sorted(b2_filtered['Aligned_Cluster'].unique()):
    clustered_data = b2_filtered[b2_filtered['Aligned_Cluster'] == cluster]
    plt.scatter(clustered_data['PA'], clustered_data['wRCPlus'], label=f'Cluster {cluster}', c=cluster_colors[cluster], alpha=0.6)
plt.title('Clustering of 2B based on PA and wRCPlus (Aligned Clusters)')
plt.xlabel('Plate Appearances (PA)')
plt.ylabel('Weighted Runs Created Plus (wRCPlus)')
plt.legend()
plt.grid(True)
plt.show()

# Display the cluster assignments with name, season, and stats
columns_of_interest = [
    'Name', 'PlayerID', 'Season', 'Age', 'G', 'PA', 'H',
    'B1', 'B2', 'B3', 'HR', 'BB', 'SO', 'AVG', 'OBP', 'SLG', 'TB', 'HBP',
    'wRCPlus', 'Bats', 'Throws', 'Height', 'Weight', 'Position', 'Aligned_Cluster'
]

```
